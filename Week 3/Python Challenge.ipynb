```python
# LEVEL 1

# this is a string
code = "g fmnc wms bgblr rpylqjyrc gr zw fylb. rfyrq ufyr amknsrcpq ypc dmp. bmgle gr gl zw fylb gq glcddgagclr ylb rfyr'q ufw rfgq rcvr gq qm jmle. sqgle qrpgle.kyicrpylq() gq pcamkkclbcb. lmu ynnjw ml rfc spj."

#this is a function
def decipher(code):
    
    #this is an empty string
    crackedCode = ''
    
    #this is a for loop, alphabet = variable name (equivalent of i in js for loop)
    for alphabet in code:
        if alphabet == 'y':
            crackedCode += 'a'
        elif alphabet == 'z':
            crackedCode += 'b'
        #for a to x
        elif ord(alphabet) >= 97 and ord(alphabet) <= 120: 
            crackedCode += chr(ord(alphabet)+2)
        #for the punctuations
        else:
            crackedCode += alphabet
    
    return crackedCode

#call function
decipher(code)

#this is a function using the recommended maketrans() method
def decipherTwo(code):
    
    #before = 'abcdefghijklmnopqrstuvwxyz'
    #after = 'cdefghijklmnopqrstuvwxyzab'
    
    conversion = str.maketrans('abcdefghijklmnopqrstuvwxyz', 'cdefghijklmnopqrstuvwxyzab')
    return(code.translate(conversion))

#call function
decipherTwo(code)

#translate "map"
decipherTwo("map")

### ANSWER = ocr

"""
NOTES:

To write a function ↓
    def functionName(parameter):
        ...
    
To write a for loop ↓
    for item in targetVariable:
        ...
        
To write a conditional ↓

    if item == thisCondition:
        ...
    elif item == thisCondition:
        ...
    else item == thisCondition:

str = built-in python function that converts value into string
ord() = converts character to interger value
chr() = converts integer value to character
maketrans() = maps out the conversion table
translate() = does the conversion/translation

-you don't need to declare variable type in Python as it automatically does it for you
-"" and '' are interchangeable
"""
```
